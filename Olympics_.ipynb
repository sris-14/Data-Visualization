```python
# Group by 'Country' and 'Gender' and count the number of occurrences
gender_counts = bios.groupby(['country', 'gender']).size().unstack(fill_value=0)


# Select the top 10 countries by some criteria, e.g., highest total players
top_countries = gender_counts.sum(axis=1).nlargest(10).index


# Calculate proportions
gender_proportions = gender_counts.div(gender_counts.sum(axis=1), axis=0)
# Filter the data to include only these top countries
gender_proportions_top = gender_proportions.loc[top_countries]


# Create a bar plot
fig, ax = plt.subplots(figsize=(7, 8))

# Plot stacked bar plot
gender_proportions_top.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])

# Set labels and title
plt.xlabel('Country')
plt.ylabel('Proportion')
plt.title('Proportion of Male and Female Players by Country')
plt.legend(title='Gender')

# Rotate x-axis labels for better readability
# plt.xticks()
n = 1  # Change this as needed
plt.xticks(ticks=np.arange(0, len(gender_proportions_top.index), n),
         rotation=45)

# Show grid for better readability
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# USA
year_data = results[results['year']>2000]
sport_count  = year_data['sport'].value_counts()

# the top 15 to 20 sports
top_n = 20
sport_count_top = sport_count.head(top_n)

print(sport_count)
plt.figure(figsize=(7,8))
# color = ['gold','silver','darkorange']
plt.pie(sport_count_top,labels=sport_count_top.index,autopct='%1.1f%%',colors=plt.get_cmap('tab10').colors)
plt.title("Proportion of sports by year(>2000)")

plt.show()
```

```python
# Filter data for years in the 1900s
data_1900s = results[(results['year'] >= 1900) & (results['year'] < 2000)]

# Filter data for years greater than 2000
data_2000s = results[results['year'] >= 2000]

# Summary statistics for each period
summary_1900s = data_1900s.describe(include='all')
summary_2000s = data_2000s.describe(include='all')

print("1900s Summary:")
print(summary_1900s)

print("\n2000s Summary:")
print(summary_2000s)
```

```python
# Drop rows with NaN values in specific columns
results_cleaned = results.dropna(subset=['year', 'country_noc', 'sport', 'medal'])

# Fill NaN values in 'medal' column with 'None' or a placeholder value
results_filled = results.fillna({'medal': 'None'})

# Remove duplicate rows based on all columns
results_no_duplicates = results_cleaned.drop_duplicates()

# Alternatively, remove duplicates based on a subset of columns (e.g., 'year', 'country', 'sport')
results_no_duplicates = results_cleaned.drop_duplicates(subset=['year', 'country_noc', 'sport'])

# Filter cleaned data for years in the 1900s
data_1900s = results_no_duplicates[(results_no_duplicates['year'] >= 1900) & (results_no_duplicates['year'] < 2000)]

# Filter cleaned data for years greater than 2000
data_2000s = results_no_duplicates[results_no_duplicates['year'] >= 2000]

# Medal counts by sport
medals_by_sport_1900s = data_1900s['sport'].value_counts()
print("medals_by_sport_1900s: \n",medals_by_sport_1900s)
medals_by_sport_2000s = data_2000s['sport'].value_counts()
print("medals_by_sport_2000s: \n",medals_by_sport_2000s)


```

```python
# Plot top sports by medal count
top_sports_1900s = medals_by_sport_1900s.head(10)
top_sports_2000s = medals_by_sport_2000s.head(10)

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
top_sports_1900s.plot(kind='bar', color='skyblue')
plt.title('Top Sports in 1900s')
plt.xlabel('Sport')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
top_sports_2000s.plot(kind='bar', color='salmon')
plt.title('Top Sports in 2000s')
plt.xlabel('Sport')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Medal counts by country
medals_by_country_1900s = data_1900s['country_noc'].value_counts()
print("medals_by_country_1900s:\n",medals_by_country_1900s)
medals_by_country_2000s = data_2000s['country_noc'].value_counts()
print("medals_by_country_2000s:\n",medals_by_country_2000s)
```

```python
# Plot top countries by medal count
top_countries_1900s = medals_by_country_1900s.head(10)
top_countries_2000s = medals_by_country_2000s.head(10)

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
top_countries_1900s.plot(kind='bar', color='brown')
plt.title('Top Countries in 1900s')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.xticks(rotation=30)

plt.subplot(1, 2, 2)
top_countries_2000s.plot(kind='bar', color='lightcoral')
plt.title('Top Countries in 2000s')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```
